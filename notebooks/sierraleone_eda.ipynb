```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file_path_benin = "../data/sierraleone-bumbuna.csv" # Adjusted path to go up one level from 'notebooks'
df = pd.read_csv(file_path_benin)
print(f"Shape of the dataset: {df.shape}")
# output some table with 5 rows
print("\nAn overview of the dataset: the first 5 rows")
print(df.head())
```

```python
print("\nAn overview of the dataset: the first 5 rows (explicitly called)")
print(df.head())
# the first 5 row
```

```python
print("\nAn overview of the dataset: a random sample of 5 rows")
print(df.sample(5))
# output random 5 row
```

```python
print("\nAn overview of the dataset: the shape and info of the dataset")
df.info()
```

```python
print("\nCheck for missing values:")
print(df.isna().sum())
```

```python
print("\nThe description of the numeric columns:")
print(df.describe())
```

```python
print("\nThe description of the categorical columns:")
print(df.describe(include=['object']))
```

```python
print("\nThe description of all columns:")
print(df.describe(include='all'))
```

```python
dup_count = df.duplicated().sum()
print("Duplicate rows:", dup_count)

cat_cols = ['Cleaning'] 
cardinality = {c: df[c].nunique() for c in cat_cols}
print("Cardinality (categoricals):", cardinality)
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()

for c in num_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[c].dropna(), kde=True) 
    plt.title(f"Distribution: {c}")
    plt.xlabel(c)
    plt.ylabel("Count")
    plt.show()
```

```python
for c in num_cols:
    plt.figure(figsize=(6, 8))
    plt.boxplot(df[c].dropna(), vert=True)
    plt.title(f"Box plot: {c}")
    plt.ylabel(c)
    plt.show()
```

```python
for c in cat_cols:
    display(df[c].value_counts().rename("count").to_frame())
```

```python
print("\nZ-score Outlier Detection:")
z = np.abs(stats.zscore(df[num_cols], nan_policy="omit"))
z_outliers_mask = (z > 3).any(axis=1) 
print("Z-score outlier rows:", int(z_outliers_mask.sum()))
```

```python

df_clean = df[~z_outliers_mask]

print("After removing Z-score outliers:", df_clean.shape)

output_path = "../data/benin_clean.csv"
df_clean.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```

```python
# Example 1: Wind Speed vs. GHI
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x="WS", y="GHI", alpha=0.6) # Using df_clean for analysis
plt.title("Wind Speed (WS) vs. Global Horizontal Irradiance (GHI)")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("GHI (W/m)")
plt.show()

# Example 2: Relative Humidity (RH) vs. Ambient Temperature (Tamb)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x="Tamb", y="RH", alpha=0.6)
plt.title("Ambient Temperature (Tamb) vs. Relative Humidity (RH)")
plt.xlabel("Ambient Temperature (C)")
plt.ylabel("Relative Humidity (%)")
plt.show()
```

```python
corr_cols = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'Tamb', 'RH', 'WS', 'BP'] # Added a few more relevant ones

corr = df_clean[corr_cols].corr()
display(corr) 

plt.figure(figsize=(10, 8)) 
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=.5)
plt.title("Correlation Matrix for Key Solar Variables")
plt.tight_layout()
plt.show()
```

```python

pairplot_cols = ['GHI', 'DNI', 'Tamb', 'RH', 'WS', 'TModA']

sns.pairplot(df_clean[pairplot_cols], diag_kind='kde') 
plt.suptitle("Pairwise Relationships of Key Solar Variables", y=1.02) 
plt.show()
```

```python

plt.figure(figsize=(7, 5))
sns.countplot(data=df_clean, x="Cleaning") 
plt.title("Count of Cleaning Events")
plt.xlabel("Cleaning Event (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
#  Step 13: Export Cleaned Data
df_clean.to_csv('../data/sierra_leone_clean.csv', index=False)
print(" Cleaned dataset exported successfully as sierra_leone_clean.csv")
```
